```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interactive

%matplotlib widget
# Define a function for a 3D plane
def ellipsoid_func_3d(a, b):
    # Create a grid of x and y values
    x = np.linspace(-10, 10, 100)
    y = np.linspace(-10, 10, 100)
    X, Y = np.meshgrid(x, y)

    # Define the plane: Z = slope_x * X + slope_y * Y
    Z = a*X**2 + b*Y**2

    Z = np.clip(Z, None, 5)

    # Set up the figure and 3D axis
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Plot the surface
    ax.plot_surface(X, Y, Z, alpha=0.5, color='red')

    # Set axis limits
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    ax.set_zlim(-10, 10)

    # Set axis labels
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()

# Create an interactive widget to vary slope_x, slope_y, and intercept
interactive_plot = interactive(ellipsoid_func_3d, a=(0.1, 2.0), b=(0.1, 2.0))
output = interactive_plot.children[-1]
interactive_plot
```
